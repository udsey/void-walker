```python
def iter_dict(obj, prefix=""):
    if isinstance(obj, dict):
        for key, value in obj.items():
            yield from iter_dict(value, f"{prefix}.{key}" if prefix else key)
    elif isinstance(obj, list):
        for i, item in enumerate(obj):
            yield from iter_dict(item, f"{prefix}[{i}]")
    else:
        yield prefix, obj
```

```python
def func(s):
    s += "!!!!!"
    return s
```

```python
def apply_to_dict(obj, func):
    if isinstance(obj, dict):
        return {key: apply_to_dict(value, func) for key, value in obj.items()}
    elif isinstance(obj, list):
        return [apply_to_dict(item, func) for item in obj]
    elif isinstance(obj, str):
        return func(obj)
    else:
        return obj  # leave int, float, bool, None unchanged
```
